## About Dataset:

### Basic Info
The Prouni - Programa Universidade para Todos (University for All Program) is an initiative created by the Brazilian government in 2005 with the objective to support students with full or partial scholarships to study in private universities.

To be able to apply for this program the student has to fulfill the following:

|  Scholarship | Family Income  |
|---|---|
|**Full Support**|  1.5 * minimum wage |
| **Partial Support** |  3 * minimum wage |

**Minimum wage in Brazil as of October 2021**:

* *BRZ 1,100.00 ~ USD 204.87*

Besides that the students must not have an university degree, must have participated in the latest Enem (national exam for federal university's application) and reach a minimum score in it.

## Reducing memory usage for each column


### Updating the 'MODALIDADE_ENSINO_BOLSA' column

**Modalidade Ensino Bolsa:**

**P**: Presencial

**D**: EAD (Educação a Distância)

```python
df3["MODALIDADE_ENSINO_BOLSA"] = df3["MODALIDADE_ENSINO_BOLSA"].replace({'PRESENCIAL':'P','EAD':'D'})
```

```python
df3["MODALIDADE_ENSINO_BOLSA"].unique()
```

### Updating the 'TIPO_BOLSA' column

**Tipo Bolsa:**

**P**: Bolsa Parcial 50%

**I**: Bolsa Integral (100%)

```python
df3["TIPO_BOLSA"].unique()
```

```python
df3["TIPO_BOLSA"] = df3["TIPO_BOLSA"].replace({'BOLSA PARCIAL 50%':'P','BOLSA INTEGRAL':'I'})
```

### This column was removed due to having over 2500 unique values

```python
del df3["NOME_CURSO_BOLSA"]
```

### Creating the 'ANO_NASCIMENTO' column

#### Only the year is required, as the day and month components are not relevant

```python
df3["DT_NASCIMENTO_BENEFICIARIO"].unique()
```

```python
df3["ANO_NASCIMENTO"] = df3["DT_NASCIMENTO_BENEFICIARIO"].str[:4].astype(int)
```

#### This column was removed as it was no longer considered relevant

```python
del df3["DT_NASCIMENTO_BENEFICIARIO"]
```

```python
df3["ANO_NASCIMENTO"].unique()
```

#### Convert the data type from int64 to int16

```python
df3["ANO_NASCIMENTO"] = df3["ANO_NASCIMENTO"].astype('int16')
```

### Updating the 'NOME_TURNO_CURSO_BOLSA' column

```python
df3["NOME_TURNO_CURSO_BOLSA"].unique()
```

**NOME_TURNO_CURSO_BOLSA**:

**N**: Noturno

**I**: Integral

**M**: Matutino

**D**: A distancia

**V**: Vespertino

```python
df3["NOME_TURNO_CURSO_BOLSA"] = df3["NOME_TURNO_CURSO_BOLSA"].replace({'Noturno':'N','Integral':'I','Matutino':'M', 'A Distancia':'D', 'Vespertino':'V'})
```

### Updating the 'RACA_BENEFICIARIO_BOLSA' column

```python
df3["RACA_BENEFICIARIO_BOLSA"].unique()
```

**RACA_BENEFICIARIO_BOLSA**:

**B**: Branca

**Pa**: Parda

**A**: Amarela

**NI**: Nao Informada

**I**: Indigena

```python
df3["RACA_BENEFICIARIO_BOLSA"] = df3["RACA_BENEFICIARIO_BOLSA"].replace({'Branca':'B', 'Parda':'Pa', 'Preta':'Pr', 'Amarela':'A', 'Nao Informada':'NI', 'Indigena':'I'})
```
